```python
# Multiples grficos en una misma figura
```

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5.0,0.1)
x2 = np.arange(0,5.0,0.2)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x1,f(x1),'ro',x2,f(x2),'k')
plt.subplot(212)
plt.plot(x2,f(x2),'g--')
plt.figure(2)
plt.plot([3,4,5])
plt.figure(1)
plt.subplot(211)
plt.title("TITULO DEL PRIMER GRAFICO")
plt.show()
```

# Textos en grficos

```python
mu = 100
sigma = 20
x = mu + sigma*np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x,50,facecolor="g",density=True,alpha=0.6)

plt.xlabel("Cociente intelectual", fontsize=13,color="green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu,\sigma)$")
plt.text(40,0.015,r'$\mu = 100,\ \sigma = 20$')
plt.axis([20,180,0,0.025])
plt.grid(True,alpha=0.2)
plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=90)
plt.subplot(1,1,1)
x = np.arange(0,10*np.pi,0.01)
y= np.cos(x)
plt.plot(x,y,lw=2.0)
plt.annotate('Mximo local',xy=(4*np.pi,1),xytext=(15,1.5),arrowprops = dict(facecolor="black",shrink= 0.05))
plt.ylim(-2,2)
plt.show()
```

```python
mu = 0.5
sd=0.3
y=mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(figsize=(10,8), dpi=90)
plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala linear")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala logaritmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y - y.mean())
plt.yscale("symlog",linthreshy=0.01)
plt.title("Plot simetrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logistica")
plt.grid(True)

plt.subplots_adjust(top=0.9,bottom=0.08,left=0.1,right=0.95,hspace=0.5,wspace=0.5)
plt.show()
```

## Ejercicio con cambios de ejes en python

```python
x = np.linspace(-np.pi,np.pi,256,endpoint=True)
S, C = np.sin(x), np.cos(x)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,S, color='blue',linewidth=1.2,linestyle='-',label="seno",zorder=.1)
plt.plot(x,C, color='green',linewidth=1.2,linestyle='-',label="coseno",zorder=.1)
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi$'])
plt.yticks(np.linspace(-1,1,3,endpoint=True),['-1',' ','+1'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.legend(loc="upper left")

x0 = 2*np.pi/3
plt.plot([x0,x0],[0,np.sin(x0)],color="blue", linewidth = 2.5, linestyle="--")
plt.scatter([x0,],[np.sin(x0),],50,color="blue")
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
             xy=(x0,np.sin(x0)),xycoords='data',xytext=(+15,+15),textcoords="offset points", fontsize=16, arrowprops=dict(arrowstyle = "->",connectionstyle="arc3,rad=.2"))

plt.plot([x0,x0],[0,np.cos(x0)],color="green", linewidth = 2.5, linestyle="--")
plt.scatter([x0,],[np.cos(x0),],50,color="green")
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
             xy=(x0,np.cos(x0)),xycoords='data',xytext=(+20,-15),textcoords="offset points", fontsize=16, arrowprops=dict(arrowstyle = "->",connectionstyle="arc3,rad=-.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="white",edgecolor="None",alpha=0.4,zorder=.2))

plt.show()
```
