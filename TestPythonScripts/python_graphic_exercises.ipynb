# Ejercicio1

```python
import matplotlib.pyplot as plt
import numpy as np
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X) 

plt.figure(figsize=(10,8))
plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.fill_between(X,Y+1,1,color="blue",alpha=.2)

plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X,Y-1,-1,where=Y-1>-1,color="blue",alpha=.2)
plt.fill_between(X,Y-1,-1,where=Y-1<-1,color="red",alpha=.2)

plt.ylim(-2.5,2.5)
plt.xlim(-np.pi,np.pi)

plt.xticks([])
plt.yticks([])

plt.show()
```

# Ejercicio 2

```python
n = 1024

X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n) 


plt.figure(figsize=(10,8))
plt.scatter(X,Y,c=np.arctan2(Y,X),cmap="jet",edgecolors="black",alpha=0.5)

plt.xticks([])
plt.yticks([])

plt.show()
```

# Ejercicio 3

```python
n = 12

X = np.arange(n)

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 

plt.figure(figsize=(10,8))
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') 

for x, y in zip(X, Y1):     
        plt.text(x, y + 0.05, '%.2f' % y, ha='center', va='bottom') 
        
for x, y in zip(X, -Y2):     
        plt.text(x, y - 0.1, '%.2f' % -y, ha='center', va='bottom') 

plt.ylim(-1.25, +1.25)

plt.xticks([])
plt.yticks([])

plt.show()
```

# Ejercicio 4

```python
Z = np.ones(20)
Z[19] = 2

explode = np.full(shape=20,fill_value=0.07)
explode[19] = 0.12

norm = plt.Normalize()
colors = plt.cm.binary(norm(np.arange(11,1,-0.5)))

plt.figure(figsize=(10,8))
plt.pie(Z,explode=explode,wedgeprops = {'edgecolor': 'black'},normalize=True, colors=colors, pctdistance=0.5)

plt.show()
```
